```python
from circuit_knitting.cutting.gate_and_wire_cutting.frontend import cut_wires_and_gates_to_subcircuits
from quantum_circuit_generator.generators import gen_uccsd
```

```python
# Create a fake observable
observables = ['ZZZZZZZZZZZZZZZZ']
```

```python

for i in range(10):

    circ = gen_uccsd(8, seed=i)

    try:
        _, _, num_wire_cuts, num_gate_cuts = cut_wires_and_gates_to_subcircuits(
            circuit=circ,
            observables=observables,
            method='automatic',
            max_subcircuit_width=10,
            max_cuts=10,
            num_subcircuits=[3],
            model='gurobi'
        )
    except ValueError:
        continue

    if num_wire_cuts != 0 and num_gate_cuts != 0:
        print('SUCCESS')
        break

```

```python
print(f'Seed: {i}')
print(num_wire_cuts, num_gate_cuts)
```
